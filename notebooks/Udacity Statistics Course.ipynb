# Lesson 4 - Bar Charts

```python
size = [1300, 1400, 1600, 1900, 2100, 2300]
price = [88000, 72000, 94000, 86000, 112000, 98000]
plt.plot(size, price, 'o')

# add a trendline
z = numpy.polyfit(size, price, 1)
p = numpy.poly1d(z)
plt.plot(size, p(size), 'r--')

plt.show()
```

```python
88000 - ((88000 - 72000) / 2)
```

```python
plt.bar(size, price, 400)
plt.show()
```

```python
income = [132754, 137192, 122177, 147121, 143000, 126010, 129200, 124312, 128132]
plt.hist(income, bins=[120000, 130000, 140000, 150000], edgecolor='black')
plt.show()
```

```python
frequency = [1003, 211, 53, 74, 336, 57, 227, 60]
work_status = [
    'Working full-time',
    'Working part-time',
    'Temporarily not working',
    'Unemployed, laid off',
    'Retired',
    'School',
    'Keeping house',
    'Other'
]
plt.title('testing')
plt.xlabel('frequency')
plt.ylabel('type')
plt.barh(work_status, frequency)
plt.xticks(rotation='vertical')
plt.gca().invert_yaxis()
plt.show()

plt.pie(frequency, labels=work_status)
plt.show()
```

# Pie Charts


```python
election_results = [50, 50]
labels = ['liberal', 'labour']
colors = ['blue', 'red']
plt.pie(election_results, labels=labels, colors=colors)
plt.show()
```

```python
party_a_votes = 724000
party_b_votes = 181000
total = party_a_votes + party_b_votes
party_a_percent = party_a_votes / total * 100
print(f'Party A got {party_a_percent}% of votes')
party_b_percent = party_b_votes / total * 100
print(f'Party B got {party_b_percent}% of votes')
```

```python
election_results = [724000, 181000]
labels = ['a', 'b']
plt.pie(election_results, labels=labels)
plt.show()
```

```python
student_ages = [12000, 96000, 36000]
labels = ['13-19', '20-32', '33-999']
plt.pie(student_ages, labels=labels)
plt.show()
```

```python
party_votes = [175000, 50000, 25000, 50000]
labels = ['A', 'B', 'C', 'D']
plt.pie(party_votes, labels=labels, startangle=90, counterclock=False)
plt.title('Votes')
plt.show()
```
